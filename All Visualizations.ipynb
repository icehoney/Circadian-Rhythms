```python
#experimenting with creating a function instead of computing each graph individually
# did not work :(
def create_graph(data):
    wb = xlrd.open_workbook(data)
    df = pd.read_excel(data)
    c = 3
    while c < 15:
        print(c)
        c += 1 
    C = df[1:].get(c)
    HR = df[1:].get('str(c+0.1)')
    A = df[1:].get('str(c+0.1)')
    plt.figure(figsize=(20, 10))
    plt.plot(C,'o-')
    plt.plot(HR)
    plt.plot(A,'r-')
    plt.title('Heart Rate, Temperature, and Activity')
```
